## Fase 2: Demostración Docente en Python
Ajustaremos un modelo utilizando la biblioteca `statsmodels`, la cual nos entrega un reporte estadístico detallado (similar al que genera R), ideal para tomar decisiones de ingeniería.

Analizaremos un caso clásico de Ingeniería Civil y Agrícola: **¿Cómo predecir la resistencia a la compresión del concreto basándonos en la cantidad de cemento utilizada en la mezcla?**

Nuestros objetivos son:
1. Leer los **Coeficientes** (Intercepto y Pendiente).
2. Evaluar el **Valor-p** de la pendiente (¿Es estadísticamente significativa la relación?).
3. Interpretar el **$R^2$** (¿Qué porcentaje de la variabilidad estamos explicando?).
4. Analizar los **Residuales** para verificar los supuestos del modelo.

#### **Celda 3: Código Python (Ajuste del Modelo y Summary)**

````
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm

sns.set_theme(style="whitegrid")
np.random.seed(123)

# 1. GENERACIÓN DEL DATASET (Simulación controlada)
# X = Cantidad de Cemento (kg/m3)
cemento_X = np.random.uniform(200, 400, 50)
# Y = Resistencia (MPa) -> Relación positiva + Error aleatorio Normal
resistencia_Y = 5.0 + 0.08 * cemento_X + np.random.normal(0, 2.5, 50)

df_concreto = pd.DataFrame({'Cemento_kg': cemento_X, 'Resistencia_MPa': resistencia_Y})

# 2. AJUSTE DEL MODELO CON STATSMODELS
# En statsmodels, debemos agregar explícitamente la constante (el intercepto Beta_0)
X_con_constante = sm.add_constant(df_concreto['Cemento_kg'])
Y = df_concreto['Resistencia_MPa']

# Entrenamos el modelo de Mínimos Cuadrados Ordinarios (OLS)
modelo = sm.OLS(Y, X_con_constante).fit()

# Mostramos el reporte completo
print(modelo.summary())

# ==========================================
# INTERPRETACIÓN INGENIERIL DEL REPORTE:
# ==========================================
# 1. Coeficientes (coef):
#    - const (Intercepto): Resistencia teórica si usáramos 0 cemento.
#    - Cemento_kg (Pendiente): Por cada kg adicional de cemento, la resistencia aumenta en ~0.08 MPa.
# 2. P>|t| (Valor-p de la pendiente): Es 0.000 (menor a 0.05). La relación es estadísticamente significativa.
# 3. R-squared: Ej. 0.75 significa que "El 75% de la variación en la resistencia del concreto
#    se explica exclusivamente por la cantidad de cemento". El 25% restante se debe a otros factores (agua, agregados).
````

#### **Celda 4: Código Python (Verificación de Supuestos mediante Residuales)**

````
# 3. ANÁLISIS DE RESIDUALES (¿Podemos confiar en el modelo?)
# Extraemos los valores predichos y los errores (residuales)
predicciones = modelo.predict(X_con_constante)
residuales = modelo.resid

fig, ax = plt.subplots(1, 2, figsize=(14, 5))

# Gráfico A: Residuales vs Predicciones (Prueba visual de Homocedasticidad)
sns.scatterplot(x=predicciones, y=residuales, ax=ax[0], color='teal', s=60)
ax[0].axhline(0, color='red', linestyle='--')
ax[0].set_title('A: Residuales vs Valores Predichos')
ax[0].set_xlabel('Resistencia Predicha (MPa)')
ax[0].set_ylabel('Residuales (Error)')

# Gráfico B: Histograma de los residuales (Prueba visual de Normalidad)
sns.histplot(residuales, kde=True, ax=ax[1], color='purple')
ax[1].set_title('B: Distribución de los Residuales')
ax[1].set_xlabel('Residuales')

plt.tight_layout()
plt.show()

# Interpretación Ingenieril:
# Gráfico A: Los puntos deben verse como una "nube aleatoria" sin forma de embudo.
# Si tienen forma de embudo (heterocedasticidad), el modelo falla en predecir resistencias altas.
# Gráfico B: El error debe seguir una distribución Normal centrada en cero.
# Si los supuestos se cumplen, nuestro modelo es seguro para tomar decisiones en la obra.
````

set.seed(123)
cemento <- runif(50, 200, 400)
resistencia <- 5.0 + 0.08 * cemento + rnorm(50, 0, 2.5)
datos_r <- data.frame(Cemento=cemento, Resistencia=resistencia)

modelo_r <- lm(Resistencia ~ Cemento, data=datos_r)
