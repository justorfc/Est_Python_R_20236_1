```markdown
## Segunda Hora: Orientación Docente en Python
La regla de oro en ingeniería probabilística es: **La distribución no se elige por la fórmula, se elige por el contexto del fenómeno.**

1. **Bernoulli:** Un solo ensayo (Éxito o Fracaso). Parámetro: $p$.
2. **Binomial:** Varios ensayos independientes. Contamos el número de éxitos. Parámetros: $n$ y $p$.
3. **Poisson:** Conteo de eventos en un intervalo continuo (tiempo, área, longitud). Parámetro: $\lambda$ (tasa promedio).

Vamos a generar datos simulados en Python usando el módulo `scipy.stats` para observar cómo se comportan estas tres distribuciones teóricas.

```

#### **Celda 3: Código Python (Distribución Bernoulli)**

````
import numpy as np
import pandas as pd
from scipy.stats import bernoulli, binom, poisson

# ==========================================
# 1. DISTRIBUCIÓN BERNOULLI
# ==========================================
print("--- 1. MODELO BERNOULLI ---")
# Contexto: Inspección de una sola pieza de acero.
# p = 0.08 (8% de probabilidad de ser defectuosa)
p_bernoulli = 0.08

# Simulamos 10 inspecciones individuales
inspecciones = bernoulli.rvs(p_bernoulli, size=10, random_state=42)
print(f"Resultados de 10 piezas (1=Defectuosa, 0=Aprobada): {inspecciones}")

# Interpretación: Bernoulli es el bloque de construcción. Solo hay dos estados posibles.
# Si repetimos esto muchas veces y sumamos los defectos, llegamos a la Binomial.
````

#### **Celda 4: Código Python (Distribución Binomial)**

````
# ==========================================
# 2. DISTRIBUCIÓN BINOMIAL
# ==========================================
print("\n--- 2. MODELO BINOMIAL ---")
# Contexto: Semillas germinadas en bandejas de 50 unidades (Ing. Agrícola).
# n = 50 semillas por ensayo, p = 0.85 (85% de probabilidad de germinación por semilla).
n_semillas = 50
p_germinacion = 0.85

# Calculamos la probabilidad teórica exacta de que germinen EXACTAMENTE 45 semillas
prob_45 = binom.pmf(k=45, n=n_semillas, p=p_germinacion)
print(f"Probabilidad de que germinen exactamente 45 semillas: {prob_45:.4f}")

# Simulamos el resultado de 5 bandejas
bandejas = binom.rvs(n=n_semillas, p=p_germinacion, size=5, random_state=42)
print(f"Semillas germinadas en 5 bandejas simuladas: {bandejas}")

# Interpretación: La Binomial requiere un "n" fijo. Sabes el máximo posible (no pueden germinar 51 semillas).
````

#### **Celda 5: Código Python (Distribución Poisson)**

````
# ==========================================
# 3. DISTRIBUCIÓN POISSON
# ==========================================
print("\n--- 3. MODELO POISSON ---")
# Contexto: Número de grietas por cada 10 metros de pavimento (Ing. Civil).
# lambda (tasa promedio) = 2.5 grietas por tramo.
lambda_grietas = 2.5

# Calculamos la probabilidad teórica de encontrar EXACTAMENTE 0 grietas (un tramo perfecto)
prob_0_grietas = poisson.pmf(k=0, mu=lambda_grietas)
print(f"Probabilidad de 0 grietas en un tramo: {prob_0_grietas:.4f}")

# Simulamos la inspección de 8 tramos de carretera
tramos = poisson.rvs(mu=lambda_grietas, size=8, random_state=42)
print(f"Número de grietas encontradas en 8 tramos simulados: {tramos}")

# Interpretación: Poisson no tiene un "límite superior fijo" evidente como la Binomial.
# Modela conteos sobre un continuo (espacio o tiempo).
````
