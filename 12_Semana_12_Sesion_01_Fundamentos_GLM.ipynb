## Segunda Hora: Orientación Docente en Python
Imagina intentar usar una línea recta para predecir la probabilidad de que un puente colapse. La línea recta eventualmente subirá por encima del 100% (probabilidad > 1) o bajará del 0% (probabilidad negativa). ¡Físicamente imposible!

Los **Modelos Lineales Generalizados (GLM)** solucionan esto utilizando una **Función de Enlace**:
1. **Regresión Logística (Enlace Logit):** Transforma la línea recta en una curva en forma de "S" (sigmoide) que siempre se mantiene entre 0 y 1. Ideal para variables binarias (Aprobado/Rechazado, Falla/No Falla).
2. **Regresión Poisson (Enlace Logarítmico):** Asegura que las predicciones de conteos nunca sean negativas. Ideal para el número de defectos, accidentes o eventos discretos.

#### **Celda 3: Código Python (El problema de la Regresión Lineal en Variables Binarias)**

````
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression

sns.set_theme(style="whitegrid")
np.random.seed(42)

# Contexto: Predecir si un motor de riego fallará (1) o no (0) basado en sus horas de uso.
horas_uso = np.linspace(100, 1000, 50)
# Simulamos la probabilidad real de falla (curva logística oculta)
prob_falla = 1 / (1 + np.exp(-(horas_uso - 550) / 80))
# Generamos los estados binarios reales (0 o 1) basados en esa probabilidad
falla_real = np.random.binomial(1, prob_falla)

df_motor = pd.DataFrame({'Horas': horas_uso, 'Falla': falla_real})

# Ajustamos una Regresión Lineal Clásica (OLS) a estos datos binarios (ESTO ES UN ERROR)
modelo_lineal = LinearRegression()
modelo_lineal.fit(df_motor[['Horas']], df_motor['Falla'])
predicciones_lineales = modelo_lineal.predict(df_motor[['Horas']])

# Visualización del problema
plt.figure(figsize=(10, 5))
plt.scatter(df_motor['Horas'], df_motor['Falla'], color='navy', label='Datos (0=Funciona, 1=Falla)')
plt.plot(df_motor['Horas'], predicciones_lineales, color='red', linewidth=2, label='Regresión Lineal Clásica')

# Líneas de imposibilidad física
plt.axhline(1, color='gray', linestyle='--')
plt.axhline(0, color='gray', linestyle='--')

plt.title('¿Por qué falla la Regresión Lineal en problemas binarios?')
plt.xlabel('Horas de Uso Continuo')
plt.ylabel('Probabilidad / Estado de Falla')
plt.legend()
plt.show()

# Interpretación Ingenieril:
# Observa la línea roja. Para motores con pocas horas (izquierda), predice probabilidades negativas (-0.2).
# Para motores muy viejos (derecha), predice probabilidades mayores al 100% (1.2).
# La regresión lineal simple NO sirve para esto. Necesitamos la curva en "S" de la Regresión Logística.
````

#### **Celda 4: Código Python (La Solución: Curva Logística)**

````
# Mostremos visualmente cómo la "Función de Enlace" arregla el problema matemático
plt.figure(figsize=(10, 5))
plt.scatter(df_motor['Horas'], df_motor['Falla'], color='navy', label='Datos (0=Funciona, 1=Falla)')

# Dibujamos la curva logística teórica (Sigmoide)
plt.plot(df_motor['Horas'], prob_falla, color='green', linewidth=3, label='Modelo Logístico (Curva "S")')

plt.title('La solución: Regresión Logística')
plt.xlabel('Horas de Uso Continuo')
plt.ylabel('Probabilidad de Falla')
plt.legend()
plt.show()

# Interpretación:
# La curva verde (Logística) respeta los límites de la física y la probabilidad.
# Nunca baja de 0 ni pasa de 1. A medida que las horas aumentan, la probabilidad
# de falla transiciona suavemente hacia el 100%. Mañana aprenderemos a ajustar
# esta curva matemáticamente usando las librerías de GLM.
````
