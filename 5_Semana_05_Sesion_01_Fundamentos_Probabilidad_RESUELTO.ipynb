## Conceptos Fundamentales de Probabilidad en Ingeniería

### Contexto en Ingeniería
En el corazón de la ingeniería, ya sea civil, agrícola o agroindustrial, reside la necesidad de tomar decisiones informadas frente a la incertidumbre. La probabilidad se convierte en nuestro lenguaje matemático esencial para cuantificar y modelar esta incertidumbre, permitiéndonos predecir resultados, evaluar riesgos y diseñar sistemas más resilientes y eficientes. Desde la estimación de la vida útil de una estructura, la predicción del rendimiento de cultivos bajo diferentes condiciones climáticas, hasta el control de calidad en procesos de fabricación, la probabilidad es una herramienta indispensable.

### Conceptos Fundamentales
*   **Experimento Aleatorio:** Cualquier proceso cuyo resultado no se puede predecir con certeza antes de su realización, pero del cual conocemos todos los posibles resultados. Por ejemplo, la inspección de una viga de concreto para determinar su estado.
*   **Espacio Muestral ($\Omega$):** El conjunto de todos los posibles resultados de un experimento aleatorio. Para la inspección de la viga, $\Omega = \{\text{Aprobado, Fallo_Menor, Fallo_Critico}\}$.
*   **Evento ($A$):** Cualquier subconjunto del espacio muestral. Un evento puede ser "la viga presenta un Fallo_Critico".

### Ecuaciones de Probabilidad
La probabilidad de un evento $A$ se denota como $P(A)$.

*   **Probabilidad Teórica (Regla de Laplace):** Cuando todos los resultados del espacio muestral son igualmente probables, la probabilidad de un evento se calcula como:
    $$
P(A) = \frac{\text{Número de resultados favorables al evento } A}{\text{Número total de resultados posibles}}
    $$

*   **Probabilidad Empírica (Frecuencia Relativa):** Basada en la observación de datos históricos o la repetición de un experimento, se calcula como:
    $$
P(A) = \frac{\text{Número de veces que el evento } A \text{ ocurre}}{\text{Número total de ensayos o observaciones}}
    $$

Ahora, vamos a llevar estos conceptos a la práctica simulando escenarios ingenieriles en Python para calcular probabilidades empíricas.

## **Celda 3: Código Python (Espacio Muestral, Eventos y Probabilidad Empírica)**

````
import pandas as pd
import numpy as np

# Simulamos el registro histórico de inspección de 500 vigas de concreto (Ing. Civil)
# o lotes de producción (Ing. Agroindustrial).
# Resultados posibles (Espacio Muestral): "Aprobado", "Fallo_Menor", "Fallo_Critico"

np.random.seed(42)
espacio_muestral = ['Aprobado', 'Fallo_Menor', 'Fallo_Critico']
# Asignamos una probabilidad teórica subyacente (80% éxito, 15% fallo menor, 5% fallo crítico)
inspecciones = np.random.choice(espacio_muestral, size=500, p=[0.80, 0.15, 0.05])

df_calidad = pd.DataFrame({'Estado_Inspeccion': inspecciones})

# ==========================================
# CÁLCULO DE PROBABILIDAD EMPÍRICA
# ==========================================
print("--- FRECUENCIAS ABSOLUTAS ---")
frecuencias = df_calidad['Estado_Inspeccion'].value_counts()
print(frecuencias)

print("\n--- PROBABILIDAD EMPÍRICA (Frecuencia Relativa) ---")
prob_empirica = df_calidad['Estado_Inspeccion'].value_counts(normalize=True)
print(prob_empirica.round(4))

# Interpretación Ingenieril:
# El "Espacio Muestral" son los 3 estados posibles.
# Un "Evento" puede ser que una viga tenga un 'Fallo_Critico'.
# La Probabilidad Empírica de 'Fallo_Critico' calculada con nuestros datos es cercana a 0.05 (5%),
# calculada simplemente dividiendo los casos favorables entre el total de inspecciones.
````

## Probabilidad Condicional e Independencia en Ingeniería

### Introducción Práctica
Tras calcular las probabilidades empíricas de eventos individuales, surge una pregunta fundamental en ingeniería: ¿Cómo cambia la probabilidad de un evento si ya sabemos que otro evento ha ocurrido? Por ejemplo, si una viga de concreto falla críticamente, ¿es esta probabilidad la misma si el material proviene del Proveedor A que si proviene del Proveedor B? La respuesta a esta pregunta nos permite entender las relaciones entre eventos y tomar decisiones más informadas, como determinar si la calidad de un material depende de su origen.

### Conceptos Clave
*   **Probabilidad Condicional:** Es la probabilidad de que un evento $A$ ocurra, *dado que* otro evento $B$ ya ha ocurrido. Nos permite actualizar nuestra creencia o medida de incertidumbre sobre $A$ a la luz de nueva información ($B$).
*   **Independencia Estadística:** Dos eventos $A$ y $B$ son estadísticamente independientes si la ocurrencia de uno no afecta la probabilidad de ocurrencia del otro. En un contexto de control de calidad, si el fallo crítico de un componente es independiente del proveedor, significa que ambos proveedores tienen tasas de fallo similares y su elección no influye en la calidad del producto final.

### Ecuaciones en LaTeX
*   La ecuación de la **Probabilidad Condicional** de $A$ dado $B$ es:

$$P(A|B) = \frac{P(A \cap B)}{P(B)}$$

donde $P(A \cap B)$ es la probabilidad de que $A$ y $B$ ocurran simultáneamente, y $P(B)$ es la probabilidad de que $B$ ocurra ($P(B) > 0$).

*   La condición matemática de **Independencia** entre $A$ y $B$ se cumple si:

$$P(A|B) = P(A)$$
o, equivalentemente:
$$P(A \cap B) = P(A) \cdot P(B)$$

### Transición al Código (Tablas de Contingencia)

Para aplicar estos conceptos, utilizaremos Pandas para construir una tabla de contingencia cruzada con nuestros datos simulados, lo que nos permitirá calcular y analizar las probabilidades condicionales entre el estado de la inspección y el proveedor.

## **Celda 4: Código Python (Probabilidad Condicional e Independencia)**

````
# Ahora, supongamos que estas inspecciones provienen de dos proveedores distintos
proveedores = np.random.choice(['Proveedor_A', 'Proveedor_B'], size=500, p=[0.6, 0.4])
df_calidad['Proveedor'] = proveedores

# Tabla de contingencia cruzada
tabla_cruzada = pd.crosstab(df_calidad['Proveedor'], df_calidad['Estado_Inspeccion'], margins=True)

print("--- TABLA DE CONTINGENCIA ---")
print(tabla_cruzada)

# Cálculo de Probabilidad Condicional
# P(Fallo_Critico | Proveedor_B) = Casos de Fallo Crítico en Prov_B / Total de Prov_B
total_prov_b = tabla_cruzada.loc['Proveedor_B', 'All']
fallos_prov_b = tabla_cruzada.loc['Proveedor_B', 'Fallo_Critico']

prob_condicional = fallos_prov_b / total_prov_b
print(f"\nProbabilidad de Fallo Crítico DADO que es del Proveedor B: {prob_condicional:.4f}")

# Interpretación: Si la probabilidad condicional es igual a la probabilidad marginal (general),
# decimos que los eventos son INDEPENDIENTES. En la práctica ingenieril, si un proveedor
# tiene una tasa de fallos mucho mayor, la calidad "depende" del proveedor.
````
