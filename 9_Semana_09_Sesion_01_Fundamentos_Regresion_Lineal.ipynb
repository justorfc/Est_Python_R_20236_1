## Segunda Hora: Orientación Docente en Python
El modelo matemático que define la regresión lineal simple es la ecuación de una línea recta, más un margen de error humano o natural:

$$Y = \beta_0 + \beta_1X + \epsilon$$

Donde:
* **$Y$**: Variable respuesta o dependiente (Lo que queremos predecir, ej. Resistencia del concreto).
* **$X$**: Variable explicativa o independiente (Lo que usamos para predecir, ej. Días de curado).
* **$\beta_0$**: Intercepto (El valor esperado de $Y$ cuando $X=0$).
* **$\beta_1$**: Pendiente (Cuánto cambia $Y$ por cada unidad que aumenta $X$).
* **$\epsilon$**: Error o residual (La diferencia entre la predicción perfecta de la línea y la realidad observada).

**Regla de oro:** La regresión modela una relación matemática, **no demuestra causalidad física**.

#### **Celda 3: Código Python (Visualizando los Componentes del Modelo)**

````
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")
np.random.seed(42)

# Contexto: Relación entre la humedad del suelo (%) y el rendimiento de un cultivo (Ton/Ha)
humedad_X = np.random.uniform(10, 30, 30) # Variable independiente (X)

# Definimos los verdaderos parámetros teóricos (Beta_0 y Beta_1)
beta_0 = 2.0  # Intercepto teórico (Rendimiento base)
beta_1 = 0.3  # Pendiente teórica (Por cada 1% de humedad, aumenta 0.3 Ton/Ha)

# Generamos el error aleatorio (epsilon) asumiendo que sigue una distribución Normal
error_epsilon = np.random.normal(loc=0, scale=1.2, size=30)

# Generamos la variable respuesta (Y) según la ecuación matemática
rendimiento_Y = beta_0 + (beta_1 * humedad_X) + error_epsilon

# ==========================================
# VISUALIZACIÓN DE LA PENDIENTE Y EL ERROR
# ==========================================
plt.figure(figsize=(10, 6))

# 1. Graficamos los puntos reales (lo que observamos en campo)
plt.scatter(humedad_X, rendimiento_Y, color='navy', s=60, label='Datos Observados')

# 2. Graficamos la "Línea Verdadera" o "Línea de Ajuste Ideal" (Sin el error)
linea_ideal = beta_0 + (beta_1 * humedad_X)
plt.plot(humedad_X, linea_ideal, color='red', linewidth=2, label='Línea de Regresión Teórica')

# 3. Dibujamos los "Residuales" (Las distancias verticales entre el punto y la línea roja)
for i in range(len(humedad_X)):
    plt.plot([humedad_X[i], humedad_X[i]], [rendimiento_Y[i], linea_ideal[i]],
             color='gray', linestyle='--', alpha=0.6)

plt.title('Componentes del Modelo: Pendiente, Intercepto y Residuales')
plt.xlabel('Humedad del Suelo (%) [X]')
plt.ylabel('Rendimiento (Ton/Ha) [Y]')
plt.legend()
plt.show()

# Interpretación Ingenieril:
# La línea roja es el modelo (Beta_0 + Beta_1*X).
# Las líneas punteadas grises son los Residuales (epsilon).
# Nuestro objetivo matemático al ajustar un modelo es encontrar la línea que
# haga que la suma de todos esos residuales (errores) sea lo más pequeña posible.
````
