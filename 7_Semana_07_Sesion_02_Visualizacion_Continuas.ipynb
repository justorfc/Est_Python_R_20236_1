## Fase 2: Demostración Docente en Python
Hoy no calcularemos probabilidades puntuales, sino que evaluaremos la **Función de Densidad de Probabilidad (PDF)**. La forma de esta curva nos dice todo sobre el fenómeno.
Analizaremos:
1. **Simetría vs. Asimetría:** Comparando la Normal (simétrica) con la Lognormal (sesgada a la derecha, ideal para caudales y tamaños de partículas).
2. **Impacto del parámetro de forma:** Veremos cómo la distribución Weibull puede cambiar radicalmente su forma para modelar fallas tempranas o desgaste por fatiga mecánica a largo plazo.

**Pregunta clave:** ¿Qué fenómeno físico estoy intentando representar con esta curva?

#### **Celda 3: Código Python (Normal vs Lognormal - Simetría vs Cola Pesada)**

````
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm, lognorm

sns.set_theme(style="whitegrid")

# ==========================================
# 1. NORMAL VS LOGNORMAL (Simetría vs Sesgo)
# ==========================================
# Generamos un eje X continuo para evaluar la densidad
x_continuo = np.linspace(0, 100, 500)

# Distribución Normal (ej. Resistencia de concreto, Media = 40, Desviación = 10)
pdf_normal = norm.pdf(x_continuo, loc=40, scale=10)

# Distribución Lognormal (ej. Tamaño de partículas de suelo, sesgo positivo)
# 's' es el parámetro de forma (desviación del logaritmo subyacente)
pdf_lognorm = lognorm.pdf(x_continuo, s=0.6, scale=40)

plt.figure(figsize=(10, 5))
plt.plot(x_continuo, pdf_normal, color='navy', linewidth=2.5, label='Normal (Simétrica)')
plt.plot(x_continuo, pdf_lognorm, color='darkorange', linewidth=2.5, label='Lognormal (Asimetría Positiva / Cola a la derecha)')

plt.title('Comparación: Distribución Normal vs Lognormal')
plt.xlabel('Valor de la variable (Ej. MPa, mm, m³/s)')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.show()

# Interpretación Ingenieril:
# La Normal asume que las desviaciones extremas son igual de probables hacia arriba que hacia abajo.
# La Lognormal nos advierte: "Cuidado, hay una pequeña probabilidad de tener valores EXTREMADAMENTE altos"
# (observa la cola naranja extendiéndose hacia la derecha).
# Ignorar esta 'cola pesada' ha causado el colapso de puentes ante caudales subestimados.
````

#### **Celda 4: Código Python (La flexibilidad de Weibull)**

````
import scipy.stats as stats

# ==========================================
# 2. EL PODER DE LA DISTRIBUCIÓN WEIBULL
# ==========================================
# Weibull es la reina de la Ingeniería de Confiabilidad porque su parámetro de forma ('c')
# le permite imitar a otras distribuciones y modelar distintas etapas de vida de un material.

x_tiempo = np.linspace(0, 3, 500)

# Escenario 1: c < 1 (Mortalidad infantil/Fallas tempranas por defectos de fábrica)
pdf_weibull_temprana = stats.weibull_min.pdf(x_tiempo, c=0.5, scale=1)

# Escenario 2: c = 1 (Tasa de fallas constante, equivale a la Exponencial)
pdf_weibull_constante = stats.weibull_min.pdf(x_tiempo, c=1.0, scale=1)

# Escenario 3: c > 1 (Fallas por desgaste o fatiga mecánica acumulada)
pdf_weibull_desgaste = stats.weibull_min.pdf(x_tiempo, c=2.5, scale=1)

plt.figure(figsize=(10, 5))
plt.plot(x_tiempo, pdf_weibull_temprana, color='red', label='Weibull c=0.5 (Falla Temprana)')
plt.plot(x_tiempo, pdf_weibull_constante, color='green', label='Weibull c=1.0 (Exponencial/Tasa Constante)')
plt.plot(x_tiempo, pdf_weibull_desgaste, color='blue', label='Weibull c=2.5 (Desgaste/Fatiga)')

plt.title('Distribución Weibull: El impacto del Parámetro de Forma (c)')
plt.xlabel('Tiempo de vida útil o ciclos de carga')
plt.ylabel('Densidad de Probabilidad (PDF)')
plt.ylim(0, 1.5)
plt.legend()
plt.show()

# Interpretación Ingenieril:
# - Línea roja: El equipo tiene fallas de fábrica y suele romperse al inicio.
# - Línea verde: El equipo falla por causas aleatorias (un impacto o sobrecarga repentina).
# - Línea azul: El equipo envejece (fatiga de materiales) y es muy predecible cuándo va a fallar.
````
