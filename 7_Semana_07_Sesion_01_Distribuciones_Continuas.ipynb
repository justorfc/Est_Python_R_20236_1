## Segunda Hora: Orientación Docente en Python
En esta etapa, la clave es conectar el modelo matemático con la realidad del campo.
* **Normal:** Errores de medición, variaciones biológicas naturales. Simétrica.
* **Lognormal:** Procesos multiplicativos, crecimiento, tamaño de partículas. Sesgada a la derecha.
* **Exponencial:** Tiempo hasta el próximo evento (ej. tiempo entre llegadas de camiones).
* **Gamma:** Suma de tiempos exponenciales (tiempo acumulado).
* **Weibull:** Confiabilidad pura. Modela desgaste progresivo o fallas tempranas en materiales y equipos.

La forma de la distribución revela la naturaleza del fenómeno. Vamos a generar datos simulados en Python usando `scipy.stats` para familiarizarnos con estos modelos.

#### **Celda 3: Código Python (Normal y Lognormal)**

````
import numpy as np
import pandas as pd
from scipy.stats import norm, lognorm, expon, gamma, weibull_min

# ==========================================
# 1. DISTRIBUCIÓN NORMAL VS LOGNORMAL
# ==========================================
print("--- NORMAL VS LOGNORMAL ---")

# Contexto Normal: Errores de medición topográfica o resistencia a compresión estándar
media_norm = 30
desviacion_norm = 5
simulacion_normal = norm.rvs(loc=media_norm, scale=desviacion_norm, size=5, random_state=42)
print(f"Resistencia simulada (Normal): {simulacion_normal.round(2)} MPa")

# Contexto Lognormal: Tamaño de partículas en un suelo o caudales extremos en un río
# Lognormal solo toma valores positivos y tiene una "cola" larga hacia la derecha.
# El parámetro 's' es la desviación estándar del logaritmo subyacente.
simulacion_lognorm = lognorm.rvs(s=0.5, scale=np.exp(3), size=5, random_state=42)
print(f"Tamaño de partículas (Lognormal): {simulacion_lognorm.round(2)} mm")

# Interpretación: Si usas la Normal para modelar caudales extremos, podrías predecir
# caudales negativos (físicamente imposible). La Lognormal corrige este problema.
````

#### **Celda 4: Código Python (Exponencial, Gamma y Weibull)**

````
# ==========================================
# 2. MODELANDO TIEMPOS Y CONFIABILIDAD
# ==========================================
print("\n--- EXPONENCIAL, GAMMA Y WEIBULL ---")

# Contexto Exponencial: Tiempo (en horas) entre fallas de un sensor de humedad (tasa constante).
# scale = 1/lambda (tiempo promedio entre fallas)
tiempo_promedio = 500
sim_expon = expon.rvs(scale=tiempo_promedio, size=5, random_state=42)
print(f"Tiempos entre fallas (Exponencial): {sim_expon.round(1)} horas")

# Contexto Gamma: Tiempo hasta que ocurren 'k' fallas (ej. vida útil hasta 3 reparaciones).
k_fallas = 3
sim_gamma = gamma.rvs(a=k_fallas, scale=tiempo_promedio, size=5, random_state=42)
print(f"Tiempo hasta {k_fallas} fallas (Gamma): {sim_gamma.round(1)} horas")

# Contexto Weibull: Vida útil de un rodamiento agroindustrial sometido a fatiga mecánica.
# Parámetro 'c' controla la forma: c>1 indica desgaste progresivo por el tiempo.
parametro_forma_c = 2.5
escala_weibull = 1000 # Vida característica
sim_weibull = weibull_min.rvs(c=parametro_forma_c, scale=escala_weibull, size=5, random_state=42)
print(f"Vida útil por fatiga (Weibull): {sim_weibull.round(1)} horas")

# Interpretación Ingenieril:
# Exponencial asume que la pieza "no envejece" (probabilidad de falla constante).
# Weibull nos permite modelar materiales que se debilitan con el uso, algo vital en Ingeniería.
````
