## Fase 2: Demostración Docente en Python

La teoría nos dice que si lanzamos una moneda justa, la probabilidad de que salga cara es del 50%. Pero si la lanzas 10 veces, ¿saldrán exactamente 5 caras? Probablemente no, debido a la variabilidad en muestras pequeñas.

Vamos a simular el fallo de un componente mecánico (Ingeniería Agrícola/Agroindustrial) o la ocurrencia de un evento climático simplificado (Ingeniería Civil). Usaremos `numpy.random` para demostrar la **Ley de los Grandes Números**: cómo la probabilidad empírica converge a la teórica a medida que aumenta el tamaño de la muestra.

```

#### **Celda 3: Código Python (Simulación y Ley de los Grandes Números)**

````
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")
np.random.seed(123)

# Simulemos un componente que tiene una probabilidad teórica de fallo del 12% (0.12)
prob_teorica_fallo = 0.12
num_simulaciones = 2000

# Generamos 2000 inspecciones aleatorias: 1 = Fallo, 0 = Aprobado
# Usamos binomial, que es equivalente a lanzar una moneda sesgada 2000 veces
inspecciones = np.random.binomial(n=1, p=prob_teorica_fallo, size=num_simulaciones)

# Calculamos la probabilidad empírica ACUMULADA paso a paso
# suma acumulada de fallos / número de intentos hasta ese punto
intentos = np.arange(1, num_simulaciones + 1)
fallos_acumulados = np.cumsum(inspecciones)
prob_empirica_acumulada = fallos_acumulados / intentos

# ==========================================
# VISUALIZACIÓN DE LA CONVERGENCIA
# ==========================================
plt.figure(figsize=(10, 5))
plt.plot(intentos, prob_empirica_acumulada, color='navy', label='Probabilidad Empírica (Simulada)')
plt.axhline(y=prob_teorica_fallo, color='red', linestyle='--', linewidth=2, label='Probabilidad Teórica (0.12)')

plt.title('Ley de los Grandes Números: Convergencia de la Probabilidad')
plt.xlabel('Número de Inspecciones (Tamaño de Muestra)')
plt.ylabel('Probabilidad de Fallo')
plt.ylim(0, 0.3)
plt.legend()
plt.show()

# Interpretación Ingenieril:
# Observa el lado izquierdo del gráfico: al principio (pocas muestras), la probabilidad estimada salta violentamente.
# Si tomas decisiones con pocas muestras, tu estimación del riesgo es muy inestable.
# Hacia la derecha (muchas muestras), la línea azul se estabiliza sobre la roja.
# Esto demuestra que a mayor cantidad de datos, más confiable es nuestra estimación del riesgo real.
````

# Configurar semilla para reproducibilidad
set.seed(123)

prob_teorica <- 0.12
n_simulaciones <- 2000

# Simular con sample(): 1 es Fallo, 0 es Aprobado
# Usamos replace=TRUE para simular extracciones independientes
inspecciones_r <- sample(c(1, 0), size=n_simulaciones, replace=TRUE, prob=c(prob_teorica, 1 - prob_teorica))

# Calcular probabilidad acumulada
intentos_r <- 1:n_simulaciones
fallos_acum_r <- cumsum(inspecciones_r)
prob_empirica_r <- fallos_acum_r / intentos_r

# Visualizar la convergencia
plot(intentos_r, prob_empirica_r, type="l", col="blue", lwd=2,
     main="Convergencia hacia la Probabilidad Teórica en R",
     xlab="Número de Inspecciones", ylab="Probabilidad Estimada", ylim=c(0, 0.3))
