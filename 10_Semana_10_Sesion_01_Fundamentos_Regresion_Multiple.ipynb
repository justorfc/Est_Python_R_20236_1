## Segunda Hora: Orientación Docente en Python
El modelo matemático se expande para incluir múltiples variables explicativas:

$$Y = \beta_0 + \beta_1X_1 + \beta_2X_2 + \dots + \beta_kX_k + \epsilon$$

* **Coeficiente Parcial ($\beta_k$):** Representa el cambio en $Y$ por cada unidad que aumenta $X_k$, **manteniendo todas las demás variables constantes** (controlando el efecto del resto).
* **Multicolinealidad:** Ocurre cuando nuestras variables explicativas ($X$) están muy correlacionadas entre sí. Esto confunde al modelo y vuelve inestables los coeficientes.
* **$R^2$ ajustado:** A diferencia del $R^2$ normal (que siempre sube si agregas más variables, aunque sean basura), el ajustado **penaliza** la inclusión de variables que no aportan valor real al modelo.

**Premisa clave:** Agregar variables no siempre mejora el modelo; a veces solo añade ruido.

#### **Celda 3: Código Python (Simulación del Fenómeno Multivariado)**

````
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")
np.random.seed(42)

# Contexto: Producción agrícola (Ton/Ha) en función de Precipitación, Temperatura y Fertilización.
n_muestras = 100

# Variables explicativas independientes (X)
precipitacion_X1 = np.random.normal(800, 150, n_muestras) # mm/año
temperatura_X2 = np.random.normal(25, 3, n_muestras)      # °C promedio
fertilizacion_X3 = np.random.normal(150, 40, n_muestras)  # kg/Ha

# Parámetros teóricos (nuestra "realidad" oculta)
beta_0 = 5.0    # Producción base
beta_1 = 0.02   # Aporte de la lluvia
beta_2 = -0.5   # Efecto negativo del calor excesivo
beta_3 = 0.05   # Aporte del fertilizante

error = np.random.normal(0, 2.5, n_muestras)

# Generamos la variable respuesta (Y)
produccion_Y = beta_0 + (beta_1 * precipitacion_X1) + (beta_2 * temperatura_X2) + (beta_3 * fertilizacion_X3) + error

df_agro = pd.DataFrame({
    'Precipitacion': precipitacion_X1,
    'Temperatura': temperatura_X2,
    'Fertilizacion': fertilizacion_X3,
    'Produccion': produccion_Y
})

# ==========================================
# VISUALIZACIÓN: MATRIZ DE CORRELACIÓN Y MULTICOLINEALIDAD
# ==========================================
# Antes de hacer la regresión, SIEMPRE debemos revisar si nuestras variables X
# compiten entre sí (Multicolinealidad).

plt.figure(figsize=(8, 6))
matriz_corr = df_agro.corr()
sns.heatmap(matriz_corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1, fmt=".2f")
plt.title('Matriz de Correlación: Explorando Multicolinealidad')
plt.show()

# Interpretación Ingenieril:
# Queremos que 'Produccion' tenga alta correlación con las demás (eso es bueno para predecir).
# PERO queremos que las variables explicativas (Precipitacion, Temperatura, Fertilizacion)
# tengan BAJA correlación entre sí. Si dos variables X tienen una correlación cercana a 1 o -1,
# tenemos multicolinealidad severa.
````

# Fijar semilla
set.seed(42)

# Simular variables
n <- 100
precipitacion <- rnorm(n, mean=800, sd=150)
temperatura <- rnorm(n, mean=25, sd=3)
fertilizacion <- rnorm(n, mean=150, sd=40)
error <- rnorm(n, mean=0, sd=2.5)

# Generar Producción (Y)
produccion <- 5.0 + (0.02 * precipitacion) + (-0.5 * temperatura) + (0.05 * fertilizacion) + error

datos_agro <- data.frame(Precipitacion=precipitacion,
                         Temperatura=temperatura,
                         Fertilizacion=fertilizacion,
                         Produccion=produccion)

# Matriz de gráficos de dispersión múltiple (Pairs plot)
# Nos permite ver visualmente la relación de todas las variables contra todas
pairs(datos_agro, main="Matriz de Dispersión Multivariada", pch=19, col=rgb(0.2,0.4,0.6,0.6))

# Correlaciones exactas
print("Matriz de Correlación Numérica:")
round(cor(datos_agro), 2)
